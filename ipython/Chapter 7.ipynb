```python
import requests
import json
import os

# you'll need to have an API key for TMDB
# to run these examples,
# run export TMDB_API_KEY=<YourAPIKey>
tmdb_api_key = os.environ["TMDB_API_KEY"]

# Setup tmdb as its own session, caching requests
# (we only want to cache tmdb, not elasticsearch)
# Get your TMDB API key from
#  https://www.themoviedb.org/documentation/api
# then in shell do export TMDB_API_KEY=<Your Key>
tmdb_api = requests.Session()
tmdb_api.params={'api_key': tmdb_api_key}

# Optional, enable client-side caching for TMDB
# Requires: https://httpcache.readthedocs.org/en/latest/
#from httpcache import CachingHTTPAdapter
#tmdb_api.mount('https://', CachingHTTPAdapter())
#tmdb_api.mount('http://', CachingHTTPAdapter())

# Some utilities for flattening the explain into something a bit more
# readable. Pass Explain JSON, get something readable (ironically this is what Solr's default output is :-p)
def flatten(l):
    [item for sublist in l for item in sublist]

def simplerExplain(explainJson, depth=0):
    result = " " * (depth * 2) + "%s, %s\n" % (explainJson['value'], explainJson['description'])
    #print json.dumps(explainJson, indent=True)
    if 'details' in explainJson:
        for detail in explainJson['details']:
            result += simplerExplain(detail, depth=depth+1)
    return result


# To speed up the pace of development, we really need to focus more heavily on the analysis and query
# settings of the search engine, rather than fidly bits of the http interface.
#
# To that end, we're going to collapse some of the code you were introduced to in chapter 3 into more general functions,
# so we can reuse them. Largely, this is the exact same code you saw in chapter 3 some more generality.

## Analyze
## The analyze function is a helper for accessing the _analyze endpoint like we did in chapter 3. Recall,
## given a field or analyzer, passing some text to _analyze will return the token stream that results from
## that analyzer. This token stream, if you recall, shows us exactly how the search engine translate text
## into individual tokens to be consumed by the underlying data structures. When we debug analysis, we see
## matches we need to expect.
def analyze(text, field=None, analyzer=None):
    whatToAnalyze = ''
    if field is not None:
        whatToAnalyze = "field=%s" % field
    elif analyzer is not None:
        whatToAnalyze = "analyzer=%s" % analyzer
    resp = requests.get("http://localhost:9200/tmdb/_analyze?%s&format=yaml" % whatToAnalyze, 
                        data=text)
    print resp.text
    
## Search
## Next we need to wrap up our execution of query DSL queries. The function 'search' will execute the passed query DSL
## query and display the results. 
## If a scoring explain is associated with the results, then it also gets displayed,
## We'll also be sure to dump the query DSL
def search(query, verbose=False):
    url = 'http://localhost:9200/tmdb/movie/_search'
    httpResp = requests.get(url, data=json.dumps(query))
    if httpResp.status_code != 200:
        print "Search Failed <%s>" % httpResp.status_code
        print "%s" % httpResp.text
    searchHits = json.loads(httpResp.text)['hits']
    print "Num\tRelevance Score\t\tMovie Title"
    for idx, hit in enumerate(searchHits['hits']):
            castNames = []            
            castCharacters = []                        
            directorNames = []
            for cast in hit['_source']['cast']:
                castNames.append(cast['name'])
                castCharacters.append(cast['character'])
            for director in hit['_source']['directors']:
                directorNames.append(director['name'])
            print "%s\t%s\t\t%s\t%s" % (idx + 1, hit['_score'], 
                                      hit['_source']['title'], 
                                      hit['_source']['vote_average'])
            if verbose:
                print "%s" % hit['_source']['title']
                print "%s" % hit['_source']['tagline']        
                print "%s" % hit['_source']['overview']        
                print "%s" % hit['_id']
                print "DIRS %s" % directorNames
                print "CAST %s" % castNames
                print "CHAR %s" % castCharacters
                if '_explanation' in hit:
                    print "%s" % simplerExplain(hit['_explanation'])
                    print "*************************************"
    
    if verbose:
        httpResp = requests.get('http://localhost:9200' + 
                    '/tmdb/movie/_validate/query?explain',
                     data=json.dumps({'query': query['query']}))
        print json.loads(httpResp.text)

## Reindex
## Reindex takes analyzer and field mappings, recreates the index, and then reindexes
## TMDB movies using the _bulk index API. There are other ways for modifying the configuration
## of the index besides dropping and restarting, however for convenience and because our data
## isn't truly that large, we'll just delete and start from scratch when we need to.
def reindex(analysisSettings, mappingSettings=None, movieDict={}):
    # Destroy any existing index (equiv to SQL "drop table")
    resp = requests.delete("http://192.168.12.5:9200/tmdb")
    print "Delete TMDB Index <%s>" % resp.status_code
    
    # Create the index with explicit settings
    # We need to explicitely set number of shards to 1 to eliminate the impact of 
    # distributed IDF on our small collection
    # See also "Relavance is Broken!"
    # http://www.elastic.co/guide/en/elasticsearch/guide/current/relevance-is-broken.html
    settings = {
        "settings": {
            "number_of_shards": 1,
            "index": {
                "analysis" : analysisSettings,
            }
        }
    }
    if mappingSettings:
        settings['mappings'] = mappingSettings
    resp = requests.put("http://192.168.12.5:9200/tmdb", data=json.dumps(settings))
    print "Create TMDB Index <%s>" % resp.status_code
    if resp.status_code != 200:
        print resp.text
    
    # Bulk index title & overview to the movie endpoint
    print "Indexing %i movies" % len(movieDict.keys())
    bulkMovies = ""
    lastMovie = None
    for id, movie in movieDict.iteritems():
        addCmd = {"index": {"_index": "tmdb", "_type": "movie", "_id": movie["id"]}}
        esDoc  = movie
        bulkMovies += json.dumps(addCmd) + "\n" + json.dumps(esDoc) + "\n"
        lastMovie = movie
    print json.dumps(lastMovie, indent=True)
    resp = requests.post("http://192.168.12.5:9200/_bulk", data=bulkMovies)
    print "Bulk Index into TMDB Index <%s>" % resp.status_code


def movieList(maxMovies=10000):
    movieIds = [];
    numPages = maxMovies / 20    
    for page in range(1, numPages + 1):
        httpResp = tmdb_api.get('https://api.themoviedb.org/3/movie/top_rated', params={'page': page})  #(1)
        if httpResp.status_code != 200:
            print "top_rated: <%s>" % httpResp.status_code
            print " why? %s " % httpResp.text
        jsonResponse = json.loads(httpResp.text) #(2)
        movies = jsonResponse['results']
        if len(movies) == 0:
            break
        for movie in movies:
            movieIds.append(movie['id'])
    return movieIds

## Extract
## Extract accesses the TMDB endpoints required and creates an entry for each movie. Recall how we iterated
## over all the movie entries at an endpoint like 'top_rated'. There's one 
## major difference between our use of TMDB here and in chapter 3: pulling more data. Not only do we access the 
## movie endpoint, we also extract the credits -- pulling in the cast (actors and such) and extracting the director.
def extract(movieIds=[], numMovies=10000):
    if len(movieIds) == 0:
        try:
            f = open('tmdb.json')
            if f:
                return json.loads(f.read());
        except IOError:
            pass    
        
    movieDict = {}
    for movieId in movieIds:
        httpResp = tmdb_api.get("https://api.themoviedb.org/3/movie/%s" % movieId)
        movie = json.loads(httpResp.text)
        
        httpResp = tmdb_api.get("https://api.themoviedb.org/3/movie/%s/credits" % movieId)
        credits = json.loads(httpResp.text)
        cast = credits['cast']
        crew = credits['crew']
        directors = []
        actors = []
        for crewMember in crew:
            if crewMember['job'] == 'Director':
                directors.append(crewMember)
        movie['cast'] = cast
        movie['directors'] = directors
        
        movieDict[movieId] = movie
    
    # Store a file locally
    with open('tmdb.json', 'w') as f:
        f.write(json.dumps(movieDict))
        f.close()
    
    return movieDict
```

# Additive vs Multiplicative Boosting

In this section, let's introduce boosting by exploring the two primary forms of boosting: additive and multiplicative. Grounded in the ability to distinguish between the two, you can then begin to bend the rules to your needs.

What are these two forms of boosting? Luckily, compared to terms like "albino elephant" these terms are relatively self explanatory:

- *Additive Boosting* -- Applies a boost by ADDING it to a set of search results
- *Multiplicative Boosting* -- Applies a boost by MULTIPLYING it to a set of search results

Why do we make such a fuss about differentiating between the two? They're both boosts that differ only by the mathematical operator that we apply between a base score and a boost. In both cases we take a *base score* -- the relevance score associated with the unboosted original query, and apply some kind of *boost* -- to the base search results. The boost can be a constant or something more complex: computed from secondary queries or formulas. We'll get to what exactly the boost should be, but first we to introduce the basics of boosting, let's compare and contrast additive and multiplicative boosting approaches.

To get back in gear, let's create a base query. A canvas for us to boost with! If you recall from the last chapter, a cross_field search over several TMDB fields did an OK job as a sort of basic relevance solution. We were finding ways to enhance it with better, more precise signals. Yet let's bring that search back as a basis for our boosting endeavours





```python
usersSearch = 'william shatner patrick stewart'
query = {
    'query': {
        'multi_match': {
           'query': usersSearch,  #User's query
            'fields': ['overview', 'title', #C
                       'directors.name', 'cast.name'],
            'type': 'cross_fields'                                
         }
    },
    'size': 5,
    'explain': True
}
search(query)
```

Still working on your Star Trek search from previous chapters, you're thinking users may often search without specifying "Star Trek. Suppose we wanted to apply a simple boost on "Star Trek" movies to help keep these near the top of the list? What would be the best way to go about it? Well based on our two boosting methods we could do one of two things:

1. Apply a healthy positive boost to Star Trek titles
2. Apply a healthy multiplier to the score of Star Trek titles

In the next two sections, let's implement a basic version of each boost form to see how it pans out.

## Layering With Additive Boosting

In this section, you'll implement a basic additive boost on Star Trek titles. What you'll see is that additive boosting has a lot in common with boolean queries you've seen throughout this book. In fact, that's precisely how they're often implemented in Elasticsearch! Additive boosts give you an opportunity to layer in additional signals to be counted during scoring. You can think of it as layering another layer on the scoring layer cake. Each layer adds a signal in the same way a Boolean query's SHOULD clause adds important criteria to the final score. To use additive boosting wisely, you need to know your ingredients! Layer it on too thick, and you'll overwhelm the other boosts and scoring signals. Layer it on too lightly and it won't hardly be noticed.

To layer in our boost just right, you need to thoroughly study your base relevance score. Remember: relevance scoring is not absolute. A score can very a great deal. Understanding how your queries perform is fundamental to additive boosting well. In this case, it appears that relevance scores for this query range up to roughly from 0.3-0.8. With that assumption, let's experiment with an additive boost for other Star Trek titles. 

First, you'll simply apply an additional should clause on top of the base query above. This isn't the only way to boost with addition, but its often the handiest!

```python
usersSearch = 'william shatner patrick stewart'
query = {
    'query': {
        'bool': {
            'disable_coord': True,
            'should': [
            {'multi_match': { #A
               'query': usersSearch,  #User's query
                'fields': ['overview', 'title', #C
                           'directors.name', 'cast.name'],
                'type': 'cross_fields'                                
             }},
            { #B
               'match_phrase': {
                        'title': 'star trek'
                    }
            }
            ]
        }
    },
    'size': 5,
    'explain': True
}
search(query)
```

In this query, there are two layers to the scoring layer cake. Each layer is a signal you need to calibrate carefully. We've already discussed the need to understand the character of the base score. But if you notice, these search results require you to also take into account the behavior of the boosting query at #B. 

There's a couple points to notice that demonstrate that the boosting query might not be working as expected

1. The scores have shot up significantly, from 0.8ish to 4 or so
2. The short "Star Trek" matches seem to outweigh the longer "Star Trek" matches

For (1) instead of a layer cake with appropriatelly equal layers, our boost layer has completely overwhelmed the base relevance score to become the dominating component of scoring. This is fixable, we can adjust this boost down for this clause.

Item (2) simply is reflecting normal TF\*IDF scoring of the Star Trek phrase. Recall that field norms bias search results towards results that are shorter. Text with "Star Trek" by itself is assumed to be proportionally more significant than text that has Star Trek plus several other words. TF\*IDF scoring reflects this assumption. Frankly, we don't care about the TF\*IDF score of the phrase "Star Trek" in our case.

You could solve (2) a number of ways. You could go about configuring this field to disable field normalization -- likely evening out the "Star Trek" scoring. Something that would be simpler would be to use Elasticsearch's `constant_score` query. The `constant_score` query simply returns the associated boost as a score instead of running the TF\*IDF scoring. 

After some experimentation, the following seems to be a better recalibration of the Star Trek boost:


```python
usersSearch = 'william shatner patrick stewart'
query = {
    'query': {
        'bool': {
            'disable_coord': True,
            'should': [
            {'multi_match': { #A
               'query': usersSearch,  #User's query
                'fields': ['overview', 'title', #C
                           'directors.name', 'cast.name'],
                'type': 'cross_fields'                                
             }},
            { #B
               'constant_score': {
                   'query': {
                       'match_phrase': {
                                'title': 'star trek'
                            }
                    },
                    'boost': 5
                }
            }
            ]
        }
    },
    'size': 5,
    'explain': True
}
search(query)
```

Additive boosting can take a lot of work to get right. For the most part, calibrating the impact of an additive boost takes the tone of calibrating any SHOULD clause in a boolean query. A great deal of testing and experimentation is required to make sure that the impact of each layer is calibrated to what you need. Without this work, its not hard for boosts to get out of whack relative to each other. Indeed, because of this difficulty many even recommend avoiding additive boosts when possible.

Nevertheless, additive boosting can be valuable in the same way that most_fields scoring from previous chapters was valuable. Remember, most_fields is a form of multi field search that adds every field's score together. Treating every field score as a signal to add to the final relevance score, you use most fields to let every underlying signal "have a vote". The more signals that come into play, the more the search engine should consider the document relevant. The same thinking comes into play here, an "additive boost" is really not much different than another signal you'd like to consider a "vote for relevance". 

When your goal is to combine these votes together, even though it may take a great deal of work, then additive boosting is for you.

# Score Skewing with Multiplicative Boosting

With additive boosting, you have to worry about carefully calibrating the impact of each layer. What if you didn't have to worry about the impact of each scoring layer? What if you simply could calibrate a multiple to apply to the score? Instead of trying to add to the scores of "Star Trek" queries, what if you just multiplied each Star Trek document by 0.5? 

The great thing about multiplicative boosting is it doesn't matter whether the base score is 0.9-1 or 90-100. The boosted documents, when multiplied by 0.5 would simply get half-again higher than the base range. So you think less with multiplicative boosting. Instead you simply tune a reasonable multiple, and amplify the scores based on the boosting signal.

For example, if instead of an additive boost for movies with Star Trek, what if we implemented a simple multiplier like the following:

```python
usersSearch = 'william shatner patrick stewart'
query = {
    'query': {
        'function_score': {
            'query': {
                 'multi_match': {
                   'query': usersSearch,  #User's query
                    'fields': ['overview', 'title', #C
                               'directors.name', 'cast.name'],
                    'type': 'cross_fields'                                
                 }
             },
             'functions': [
                {
                     'weight': 2.5,
                     'filter': {
                        'query': {
                             'match_phrase': {
                                    'title': 'star trek'
                            }
                        }
                     }
                 }
            ]
        }
    },
    'size': 50,
    'explain': True
}
search(query, verbose=True)
```
